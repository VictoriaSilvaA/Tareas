```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## se importan los datos
pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/pobXedadCenso.csv")
pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")

data1 = pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
data2 = pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/pobXedadCenso.csv")
vacunas = pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")
```

```python
RegionM=vacunas[vacunas['Region']=='Metropolitana']
RegionM
```

```python
col_list=list(RegionM)

col_list.remove('Poblacion')
col_list.remove('Region')
col_list.remove('Comuna')
col_list.remove('Codigo region')
col_list.remove('Codigo comuna')

RegionM=vacunas[vacunas["Region"].str.contains("Metropolitana")]
RegionM=RegionM.drop([140],axis=0)

RegionM["Total"] = RegionM[col_list].sum(axis=1)
RegionM
```

```python
RegionM=RegionM[['Region','Comuna','Codigo comuna','Poblacion','Total']]
RegionM

## Se eliminaron las columnas que no nos sirve para el analisis y se dejaron las que si importaban
```

```python
data2['EDAD_65YMAS'] = data2['EDAD_65YMAS'].replace('*', 0)

data2['EDAD_15A64'] = data2['EDAD_15A64'].replace('*', 0)

data2['EDAD_6A14'] = data2['EDAD_6A14'].replace('*', 0)

data2['EDAD_0A5'] = data2['EDAD_0A5'].replace('*', 0)

data2

## se reemplazaron los datos que no tenian valor por 0 y asi poder trabajarlo
```

```python
type(data2['EDAD_0A5'][4])

data2['EDAD_65YMAS']= pd.to_numeric(data2['EDAD_65YMAS'])
data2['EDAD_15A64']= pd.to_numeric(data2['EDAD_15A64'])
data2['EDAD_6A14'] = pd.to_numeric(data2['EDAD_6A14'])
data2['EDAD_0A5'] = pd.to_numeric(data2['EDAD_0A5'])

data1.rename(columns = {'comuna' : 'Codigo comuna'}, inplace = True)
data2.rename(columns = {'COMUNA' : 'Codigo comuna'}, inplace = True)
data2

## Se pasa cada columna a numero y asi poder trabajarla
```

```python
data2=data2.groupby('Codigo comuna').sum()

## se suman las cantidad de personas de cada rango de edad y codigo de comuna
```

```python
Q=pd.merge(RegionM,data2,
how='inner', on='Codigo comuna')
Q
## se combinan ambos datos y la nombramos Q
```

```python
data1.rename(columns = {'comuna' : 'Codigo comuna'}, inplace = True)
```

```python
RegionM
```

```python
W= pd.merge(Q,data1,how='inner', on='Codigo comuna')
W

## se combinan los archivos data1 y Q y la nombramos W
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.scatter(x = W.Total/W.Poblacion, y = W.HDI, alpha = 0.8, color ='red')
ax.set_xlabel('Porcentaje_Vacunados')
ax.set_ylabel('HDI')

## se grafican ahora los datos ya trabajados del porcentaje de vacunados de cada comuna de la RM respecto a su HDI
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.scatter(x = W.Total/W.Poblacion, y = W.EDAD_65YMAS/W.Poblacion, alpha= 0.8, color ='green')
ax.set_xlabel('Porcentaje_Vacunados')
ax.set_ylabel('Porcentaje_de_adultomayor')

## porcentaje de vacunados de cada comuna con la cantidad de adultos mayor
```
