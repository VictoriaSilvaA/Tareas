```python
Data_1 = open('FallecidosEtario_T.csv')
datos = pd.read_csv(Data_1, header=0)
    
```

```python
Lista=[]                   # muertes por covid a la semana
datos = datos.values
Fechas =[]
Muertes = []
Date=[]
Death=[]
n=0
Casos_semana=0

for row in datos:
    n=n+1
    x=row[0]
    y=int(row[1])+int(row[2])+int(row[3])+int(row[4])+int(row[5])+int(row[6])+int(row[7])
    Fechas.append(x)
    Muertes.append(y)
    if n/7==1:
        Date.append(x)
        Death.append(y)
        n=0
        Casos_semana=0

fig, ax = plt.subplots()

ax.plot(Date,Death, linestyle = "solid")
plt.show()
```

```python
Death_T = []
a = 0

for j in range(len(Death)):
    a = Death[j]-Death[j-1]
    Death_T.append(a)

Death_T[0] = Death[0]

fig, ax = plt.subplots()
ax.plot(Date, Death_T, linestyle = "solid")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
Lista = [] # codigos para filtrar y generar muertes totales a la semana
with open('Def.csv', newline='') as File:
    datos_2 = csv.reader(File, delimiter=';')
    
    for row in datos_2:
        Lista.append(row[1])
        Lista.append(row[12])
```

```python
Cantidad = []
for f in Fechas:
    datos = Lista.count(f)
    Cantidad.append(datos)
```

```python
fallecidos_semana = 0
n = 0
Fallecidos = []

for c in Cantidad:
    n = n + 1
    fallecidos_semana = fallecidos_semana + c
    
    if n/7 == 1:
        Fallecidos.append(fallecidos_semana)
        n = 0
        fallecidos_semana = 0
```

```python
fig, ax = plt.subplots()  #muertes totales a la semana

ax.plot(Date,Fallecidos, linestyle = "solid")
plt.show()
```

```python
Muertes_Gripe = []           # muertes por gripe a la semana

for j in range(len(Lista)):
    
    if Lista[j] == "Influenza [gripe] y neumonía":
        Muertes_Gripe.append(Lista[j-1])
```

```python
Cantidad_Gripe = []
for f in Fechas:
    cant = Muertes_Gripe.count(f)
    Cantidad_Gripe.append(cant)
```

```python
semana = 0
n = 0
Fallecidos_Gripe = []

for c in Cantidad_Gripe:
    n = n + 1
    semana = semana + c
    
    if n/7 == 1:
        Fallecidos_Gripe.append(semana)
        n = 0
        semana = 0
```

```python
fig, ax = plt.subplots()  #muertes totales a la semana por gripe

ax.plot(Date,Fallecidos_Gripe, linestyle = "solid")
plt.show()
```

```python
fig, ax = plt.subplots()                          
ax.plot(Date, Fallecidos ,linestyle="solid")
ax.plot(Date, Death_T, linestyle = "dotted")
plt.show()
```
